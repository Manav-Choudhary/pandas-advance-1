```python
#Ques 2
import pandas as pd
df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
df.reset_index(drop=True,inplace=True)
def func(df):
    new_index=range(1,2*len(df),2)
    new_df=df.copy()
    new_df.reset_index(drop=True,inplace=True)
    new_df.index=new_index
    return new_df
new_df=func(df)
new_df.head()
```

```python
#Ques 7
import pandas as pd
def func(email):
    return email.split('@')[0]
sample_data = [
    'john@example.com',
    'jane@example.com',
    'smith@example.com',
    'mike@example.com',
    'alice@example.com',
    'bob@example.com',
    'emma@example.com',
    'peter@example.com',
    'susan@example.com',
    'alex@example.com'
]
data={
    "email_id":sample_data
}
df=pd.DataFrame(data)
df['User-Name']=df['email_id'].apply(func)
df
```

```python
#Ques 8
import pandas as pd

def func(df):
    return df[(df['A'] > 5) & (df['B'] < 10)]

data = {
    'A': [3, 8, 6, 2, 9],
    'B': [5, 2, 9, 3, 1],
    'C': [1, 7, 4, 5, 2]
}

df = pd.DataFrame(data)
df = func(df)

print(df)
```

```python
#Ques 10
import pandas as pd
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05', '2023-01-06', '2023-01-07', '2023-01-08'],
    'Sales': [100, 120, 90, 80, 110, 150, 130, 140]
}
def func(df):
    df['Moving_avg']=df['Sales'].rolling(window=1).mean()
    return df
df=pd.DataFrame(data)
new_df=func(df)
new_df.head()
```

```python
#Ques 11
import pandas as pd
def func(date):
    date_obj=pd.to_datetime(date)
    return date_obj.strftime('%A')
data={
    'Date':['2023-01-01','2023-01-02','2023-01-03','2023-01-04','2023-01-05']
}
df=pd.DataFrame(data)
df['Weekday']=df['Date'].apply(func)
df
```
